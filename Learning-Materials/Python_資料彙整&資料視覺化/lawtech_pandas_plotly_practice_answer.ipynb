```python
import pandas as pd
data = pd.read_csv("LawTech_0727_Pandas.csv",index_col=0) 
data.head()
```

```python
location = []
year = []
for i in data["title"]:
    a = i.split(" ")
    location.append(a[0])
    year.append(a[1])
```

```python
data["year"] = year
data["location"] = location
```

```python
#練習1:取判決編號以及判決類別
number = []
judge_type = []
for i in data["title"]:
    a = i.split(" ")
    number.append(a[5])
    b = i.split("號")[1].split("（")
    judge_type.append(b[0])
```

```python
data["number"] = number
data["judge_type"] = judge_type
data.head(3)
```

```python
#練習2:以loc及iloc方式取出第15~20筆資料且只有呈現他的日期(date)及地點(location)
data.loc[15:20,["date","location"]]
```

```python
data.iloc[15:21,[1,6]]
```

```python
#練習3:取出年份為104年且判決為"違反證券交易法等罪"的資料 
data.groupby("year").get_group("104").groupby("judge").get_group("違反證券交易法等罪")
```

```python
#練習4:篩選出日期為下半年度的資料(6~12月)，且只有顯示標題(title)、日期(date)、內容(content)
month = []
def get_year(data):
    return int(data["date"].split(".")[1])
data["month"] = data.apply(get_year,axis=1)

a= data[data["month"]>=6]
a.filter(["title","date","content"])
```
